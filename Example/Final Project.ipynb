```python
transformed_dataset = dataset.drop(['body','cabin','boat'], axis=1)
transformed_dataset.count()
```

```python
transformed_dataset['home.dest'] = transformed_dataset['home.dest'].fillna('NA')
transformed_dataset = transformed_dataset.dropna()
transformed_dataset.count()
```

```python
transformed_dataset.head()
```

```python
def preprocess_df(df):
    PPDf = df.copy()
    encoder = preprocessing.LabelEncoder()
    PPDf.sex = encoder.fit_transform(PPDf.sex)
    PPDf.embarked = encoder.fit_transform(PPDf.embarked)
    
    PPDf = PPDf.drop(['name', 'ticket', 'home.dest'], axis=1)
    return PPDf
```

```python
processed_df = preprocess_df(transformed_dataset)
preprocess_df.he
```

```python
# Split-out validation dataset into the features(X) and target(Y)
X = processed_df.drop(['survived'], axis=1).values
Y = processed_df['survived'].values
```

```python
validation_size = 0.20

X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=7)
```

```python
#Test Harness
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))


results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=7)
    
    cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
    
    results.append((name ,cv_results))
    names.append(name)
    msg = "%s Mean Results of 10 splits: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
plt.pyplot.figure(figsize=(20,5))
for result in results:
    PerformancePlot = plt.pyplot.plot(result[1], label=result[0])
    plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
